```python
def sacarComas(numero):
    try:
        if(numero < 2 and numero > 1):
            return round(numero * 1000)
        if(numero < 3 and numero > 2):
            return round(numero * 1000)
        if(numero < 4 and numero > 3):
            return round(numero * 1000)
    except:
        pass
    return numero

import numpy as np
def limpiarEspacio(texto):
    try:
        return texto.strip()
    except:
        return np.nan
```

```python
def getNumber(numero):    
    try:
        numero = str(numero).replace("%","").replace(",",".")
        numero = float(numero)
        return numero 
    except:
        return 0
    
df = pd.concat(acumulador)
df["semmana_interna"] = df["semmana_interna"].round().astype(int).astype(str)
df["nombre final"] = df.apply(lambda x: f"{x.nombre} {x.nombre1} {x.nombre2}".replace("nan",""),axis=1)
del df["nombre"]
del df["nombre1"]
del df["nombre2"]
df["semana"] = df["semmana_interna"].apply(lambda x:variable_cargada[x])
df["semana"] = df["semmana_interna"].apply(lambda x:variable_cargada[x])
df["FechaInicial"] = df["semana"].apply(lambda x:datetime.strptime(x.split(" ")[3], "%d/%m/%Y"))
df["FechaFinal"]   = df["semana"].apply(lambda x:datetime.strptime(x.split(" ")[5], "%d/%m/%Y)"))
df["actual"] = 2023
df["anterior"] = 2022

df["nombre"] = df["nombre final"]
del df["nombre final"]

del df["semmana_interna"]
df["Delito"] = df["Delito"].apply(lambda x: x.strip())
#df["ultimos_7_dias_var"] = df["ultimos_7_dias_var"].apply(getNumber)
#df["ultimos_28_dias_var"] = df["ultimos_28_dias_var"].apply(getNumber)
#df["anno_a_la_fecha_var"] = df["anno_a_la_fecha_var"].apply(getNumber)

for i in ['ultimos_7_dias_anterior', 'ultimos_7_dias', 'ultimos_7_dias_abs',
       'ultimos_7_dias_var', 'ultimos_28_dias_anterior', 'ultimos_28_dias',
       'ultimos_28_dias_abs', 'ultimos_28_dias_var',
       'anno_a_la_fecha_anterior', 'anno_a_la_fecha', 'anno_a_la_fecha_abs',
       'anno_a_la_fecha_var']:
    df[i] = df[i].apply(getNumber)    

merge = df.merge(carabinero,left_on="COD_CUADRANTE_INTERNO",right_on="CUA_CODIGO", how="outer")
merge["titulo"] = merge["CUA_DESCRI"]
merge["COM_CODIGO"] = merge.apply(getComunasAparte,axis=1)
merge2 = merge.rename(columns=diccionarioColumnas)[columnasRef]
diccionarioComunas = {}
for j,i in merge2[["COM_CODIGO","COMUNA"]].drop_duplicates().dropna().iterrows():
    diccionarioComunas[i["COM_CODIGO"]] = i["COMUNA"]

def getComunasAparte2(fila):
    try:
        return diccionarioComunas[int(fila["COM_CODIGO"])]
    except:
        return fila["COMUNA"]
merge2["COMUNA"] = merge2.apply(getComunasAparte2,axis=1)

merge2["año"] = merge2["año"].apply(sacarComas)
merge2["año ant"] = merge2["año ant"].apply(sacarComas)
merge2["año var"] = merge2["año var"].apply(sacarComas)
merge2['mes'] = merge2['mes'].apply(sacarComas)
merge2['mes ant'] = merge2['mes ant'].apply(sacarComas)
merge2['mes var'] = merge2['mes var'].apply(sacarComas)
#merge2['delito'] = merge2['delito'].apply(limpiarEspacio)
listaSemana = list(merge2['id_semana'].unique())[:-1]
merge3 = pd.concat([merge2])
#base2 = base[base["id_semana"].apply(lambda x: x not in listaSemana)]
 
#merge3 = pd.concat([merge2,base2])
merge3['delito'] = merge3['delito'].apply(limpiarEspacio)
merge3.to_excel(r"C:\Users\limc_\carabineros\robot3_valpo\consolidado_noumbrales_valpo_2023.xlsx", index=False)
print("Termino")
```
