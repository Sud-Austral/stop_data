```python
acumulador = []
df2 = df[['semana ant','año ant','id_semana', 'delito', 'informacion','CUA_CODIGO',"COMUNA",'UNIDAD', 'DESTACAMEN','PREFECTURA',"CUA_DESCRI"]]
df2["Año Diferencia"] = df2["año ant"] - df2["semana ant"]
#df2 = df2[df2["semana ant"].notnull()]
df2["semana ant"] = df2["semana ant"].fillna(0)

df2 = df2.sort_values(by=['CUA_CODIGO', 'id_semana','delito', 'informacion'], ascending=[True, False,False,False])
for i,j in df2[['delito', 'informacion','CUA_CODIGO']].drop_duplicates().iterrows():
    delito_param = j["delito"]
    informacion_param = j["informacion"]
    CUA_CODIGO_param =  j["CUA_CODIGO"]
    print(CUA_CODIGO_param)
    query_1 = df2.query("delito == @delito_param and informacion == @informacion_param and CUA_CODIGO == @CUA_CODIGO_param")
    query_1 = query_1.sort_values(by=['id_semana'], ascending=False)
    query_1['Año Diferencia2'] = query_1['Año Diferencia'].shift(1)
    query_1["verificacion"] = query_1.apply(lambda x: x["año ant"] == x["Año Diferencia2"], axis=1)
    acumulador.append(query_1.copy())
    
    #print(j)    1
#df2
```

```python
pd.concat(acumulador).to_excel("verificacion6.xlsx", index=False)
```
