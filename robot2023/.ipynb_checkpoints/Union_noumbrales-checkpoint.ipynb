```python
"""
columnasRef = ['semana', 'semana ant', 'semana var', 'semana var%', 'mes', 'mes ant',
       'mes var', 'mes var%', 'año', 'año ant', 'año var', 'año var%',
       'delito', 'informacion', 'id_semana', 'actual-año', 'anterior-año',
       'Cuad_titulo', 'CUA_CODIGO', 'type', 'features', 'CUA_SUP', 'COMUNA',
       'UNIDAD', 'DESTACAMEN', 'PREFECTURA', 'ZONA', 'NUM_CUAD', 'CUA_DESCRI',
       'CUA_TIPO', 'CUA_ESTADO', 'CUA_FTERMI', 'UNI_CODIGO', 'CUA_ANO',
       'COD_AUPOL', 'COM_CODIGO', 'SHAPE_LENG', 'SHAPE_AREA', 'COD_DESTAC',
       'TOOLTIP', 'COD', 'SUP_KM2', 'FechaInicial', 'FechaFinal',
       'COD_CUADRANTE_INTERNO']
"""
columnasRef = ['semana', 'semana ant', 'semana var', 'semana var%', 'mes', 'mes ant',
       'mes var', 'mes var%', 'año', 'año ant', 'año var', 'año var%',
       'delito', 'informacion', 'id_semana', 'actual-año', 'anterior-año',
       'Cuad_titulo', 'CUA_CODIGO', 'COMUNA',
       'UNIDAD', 'DESTACAMEN', 'PREFECTURA', 'ZONA', 'NUM_CUAD', 'CUA_DESCRI',
       'CUA_TIPO', 'CUA_ESTADO', 'CUA_FTERMI', 'UNI_CODIGO', 'CUA_ANO',
       'COD_AUPOL', 'COM_CODIGO', 'COD_DESTAC',
       'FechaInicial', 'FechaFinal',
       'COD_CUADRANTE_INTERNO']
diccionarioColumnas = {
                        'ultimos_7_dias_anterior':'semana ant',
                        'ultimos_7_dias':'semana', 
                        'ultimos_7_dias_abs':'semana var',
                        'ultimos_7_dias_var':'semana var%', 
                        'ultimos_28_dias_anterior': 'mes ant', 
                        'ultimos_28_dias':'mes',
                        'ultimos_28_dias_abs':'mes var', 
                        'ultimos_28_dias_var':'mes var%',
                        'anno_a_la_fecha_anterior':'año ant', 
                        'anno_a_la_fecha':'año', 
                        'anno_a_la_fecha_abs':'año var',
                        'anno_a_la_fecha_var':'año var%', 
                        'Delito':'informacion',
                        'nombre':'delito',
                        'semana':'id_semana',
                        'actual':'actual-año',
                        'anterior':'anterior-año',
                        'titulo':'Cuad_titulo'
                        }

web = pd.read_excel(r"C:\Users\limc_\carabineros\web_cuadrantes.xlsx",sheet_name="aparte")
diccionarioWeb = {}
for j,i in web[["codigo_cuadrante","codigo_comuna"]].iterrows():
    diccionarioWeb[i["codigo_cuadrante"]] = i["codigo_comuna"]

def getComunasAparte(fila):
    try:
        return diccionarioWeb[int(fila["COD_CUADRANTE_INTERNO"])]
    except:
        return fila["COM_CODIGO"]    

    
diccionarioWeb    
```

```python
# Ruta de la carpeta que deseas recorrer
carpeta = 'noumbrales'

# Lista para almacenar los nombres de los archivos Excel encontrados
archivos_excel = []

# Recorre la carpeta
for filename in os.listdir(carpeta):
    if filename.endswith('.xlsx') or filename.endswith('.xls'):
        archivos_excel.append(os.path.join(carpeta, filename))
        
acumulador = []

# Ahora, puedes leer los archivos Excel
for archivo_excel in archivos_excel:
    try:
        print(archivo_excel)
        df = pd.read_excel(archivo_excel)
        del df['Unnamed: 0']
        del df['index']
        df.columns = columnas
        df = df.drop(index=[0, 1]).reset_index(drop=True)
        df["semmana_interna"] = int(archivo_excel[29:-5])
        acumulador.append(df.copy())
        #df
        # Haz algo con el DataFrame, por ejemplo, imprimir las primeras filas
        #print(f'Contenido de {archivo_excel}:')
        #print(df.head())
    except Exception as e:
        print(f'Error al leer {archivo_excel}: {str(e)}')
```

```python
def sacarComas(numero):
    try:
        if(numero < 2 and numero > 1):
            return round(numero * 1000)
        if(numero < 3 and numero > 2):
            return round(numero * 1000)
        if(numero < 4 and numero > 3):
            return round(numero * 1000)
    except:
        pass
    return numero
```

```python
def getNumber(numero):    
    try:
        numero = str(numero).replace("%","").replace(",",".")
        numero = float(numero)
        return numero 
    except:
        return 0
    
df = pd.concat(acumulador)
df["semmana_interna"] = df["semmana_interna"].round().astype(int).astype(str)
df["nombre final"] = df.apply(lambda x: f"{x.nombre} {x.nombre1} {x.nombre2}".replace("nan",""),axis=1)
del df["nombre"]
del df["nombre1"]
del df["nombre2"]
df["semana"] = df["semmana_interna"].apply(lambda x:variable_cargada[x])
df["semana"] = df["semmana_interna"].apply(lambda x:variable_cargada[x])
df["FechaInicial"] = df["semana"].apply(lambda x:datetime.strptime(x.split(" ")[3], "%d/%m/%Y"))
df["FechaFinal"]   = df["semana"].apply(lambda x:datetime.strptime(x.split(" ")[5], "%d/%m/%Y)"))
df["actual"] = 2023
df["anterior"] = 2022

df["nombre"] = df["nombre final"]
del df["nombre final"]

del df["semmana_interna"]
df["Delito"] = df["Delito"].apply(lambda x: x.strip())
#df["ultimos_7_dias_var"] = df["ultimos_7_dias_var"].apply(getNumber)
#df["ultimos_28_dias_var"] = df["ultimos_28_dias_var"].apply(getNumber)
#df["anno_a_la_fecha_var"] = df["anno_a_la_fecha_var"].apply(getNumber)

for i in ['ultimos_7_dias_anterior', 'ultimos_7_dias', 'ultimos_7_dias_abs',
       'ultimos_7_dias_var', 'ultimos_28_dias_anterior', 'ultimos_28_dias',
       'ultimos_28_dias_abs', 'ultimos_28_dias_var',
       'anno_a_la_fecha_anterior', 'anno_a_la_fecha', 'anno_a_la_fecha_abs',
       'anno_a_la_fecha_var']:
    df[i] = df[i].apply(getNumber)    

merge = df.merge(carabinero,left_on="COD_CUADRANTE_INTERNO",right_on="CUA_CODIGO", how="outer")
merge["titulo"] = merge["CUA_DESCRI"]
merge["COM_CODIGO"] = merge.apply(getComunasAparte,axis=1)
merge2 = merge.rename(columns=diccionarioColumnas)[columnasRef]
diccionarioComunas = {}
for j,i in merge2[["COM_CODIGO","COMUNA"]].drop_duplicates().dropna().iterrows():
    diccionarioComunas[i["COM_CODIGO"]] = i["COMUNA"]

def getComunasAparte2(fila):
    try:
        return diccionarioComunas[int(fila["COM_CODIGO"])]
    except:
        return fila["COMUNA"]
merge2["COMUNA"] = merge2.apply(getComunasAparte2,axis=1)

merge2["año"] = merge2["año"].apply(sacarComas)
merge2["año ant"] = merge2["año ant"].apply(sacarComas)
merge2["año var"] = merge2["año var"].apply(sacarComas)
merge2['mes'] = merge2['mes'].apply(sacarComas)
merge2['mes ant'] = merge2['mes ant'].apply(sacarComas)
merge2['mes var'] = merge2['mes var'].apply(sacarComas)

#merge2.to_excel("consolidado_noumbrales.xlsx", index=False)
```

```python
merge2.to_excel("consolidado_noumbrales.xlsx", index=False)
print("Hola")
```
